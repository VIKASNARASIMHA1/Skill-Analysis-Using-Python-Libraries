```python
import ast
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt  
dataset = pd.read_csv('data_jobs.csv')
```

```python
dataset['job_posted_date'] = pd.to_datetime(dataset['job_posted_date'])
dataset['job_skills'] = dataset['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
## Skill count per month for data analysts
df_ind = dataset[(dataset['job_title'] == 'Data Analyst') & (dataset['job_country'] == 'India')].copy()

df_ind['job_posted_month_no'] = df_ind['job_posted_date'].dt.month

# Explode job_skills column
df_explode = df_ind.explode('job_skills')
```

```python
df_explode
```

```python
df_pivot = df_explode.pivot_table(index='job_posted_month_no', columns='job_skills',  aggfunc='size', fill_value = 0)
```

```python
df_pivot
```

```python
df_pivot.loc['Total'] = df_pivot.sum()
df_pivot = df_pivot[df_pivot.loc['Total'].sort_values(ascending=False).index]
df_pivot = df_pivot.drop('Total')

df_pivot
```

```python
df_pivot.iloc[:, :5].plot(kind='line')

plt.title('Trending Top Skills for Data Analysts in the US')
plt.ylabel('Count')
plt.xlabel('')
plt.show()
```

```python
DA_totals = df_ind.groupby('job_posted_month_no').size()

DA_totals
```

```python
df_ind_perc = df_pivot.iloc[:12].div(DA_totals/100, axis=0)

# changes month number to month name
df_ind_perc = df_ind_perc.reset_index()
df_ind_perc['job_posted_month'] = df_ind_perc['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_ind_perc = df_ind_perc.set_index('job_posted_month')
df_ind_perc = df_ind_perc.drop(columns='job_posted_month_no')

df_ind_perc
```

```python
from matplotlib.ticker import PercentFormatter

df_plot = df_ind_perc.iloc[:, :5]
sns.lineplot(data=df_plot, dashes=False, legend='full', palette='tab10')
sns.set_theme(style='ticks')
sns.despine() # remove top and right spines

plt.title('Trending Top Skills for Data Analysts in the US')
plt.ylabel('Likelihood in Job Posting')
plt.xlabel('2023')
plt.legend().remove()
plt.gca().yaxis.set_major_formatter(PercentFormatter(decimals=0))

# annotate the plot with the top 5 skills using plt.text()
for i in range(5):
    plt.text(11.2, df_plot.iloc[-1, i], df_plot.columns[i], color='black')

plt.show()
```
